```python
import pyspark
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder \
    .master("local[*]") \
    .appName('test') \
    .getOrCreate()
```

```python
COVID_DATA_PATH = '../data/covid_19_data.csv'
DEATH_DATA_PATH = '../data/time_series_covid_19_deaths.csv'
```

```python
df = spark.read.option('header', 'true').csv(COVID_DATA_PATH)
```

```python
from pyspark.sql import types
```

```python
schema = types.StructType([
    types.StructField('SNo', types.IntegerType(), True), 
    types.StructField('ObservationDate', types.StringType(), True), 
    types.StructField('Province/State', types.StringType(), False), 
    types.StructField('Country/Region', types.StringType(), True), 
    types.StructField('Last Update', types.StringType(), True), 
    types.StructField('Confirmed', types.FloatType(), True), 
    types.StructField('Deaths', types.FloatType(), True), 
    types.StructField('Recovered', types.FloatType(), True)
])
```

```python
df = spark.read \
    .option("header", "true") \
    .schema(schema) \
    .csv(COVID_DATA_PATH)
```

```python
df.head(10)
```

```python
df = df.filter(df['Country/Region'] == 'Mainland China')
```

```python
df.write.parquet('../data/records/', mode='overwrite')
```
